# Regression with neural networks

Note: This ipython notebook (and the others of the serie) is part of [this project](http://www.marcodena.it/blog/detecting-anomalies-with-neural-newtorks/). In order to have a full description, I suggest you to read the paper. Here only the "operative" commands are explained.

As stated before, the system is composed by a 2-phase process where in the first one the short-term (hourly) gas consumption is forecasted thanks to a historical time-series, and in the second phase outliers are detected on the basis of deviations from
expected (or forecast).

Without a good model, the objective to detect outliers will not be achieved. Before explaining the Neural network, let's have a look on the data and engineering the features.

The ANN has been trained with early stopping, with a fixed number of training epochs (phases) or stopping the training when the validation error rate was increasing. The network is always trained with $70\%$ of the data, $15\%$ is used for validation and the other $15\%$ for testing.

(all the details are in the paper)

```python
import pandas as pd
import numpy as np
building = '740-NTH' #'882-WBW' #
df = pd.read_csv('datasets/'+building+'_elektriciteit_withWeather_clean.csv')
df['Momentum'] = df['Momentum'].astype('datetime64')
df = df.set_index('Momentum', drop=False)

# There is an implicit conversion of timezone. I need to fix it.
df.index = df.index + pd.DateOffset(hours=1)
del df['Momentum']
df['Momentum'] = df.index
df.head()

df.head()
```

The hour is the first feature to add.

```python
df['hour'] = 0
df['sinHour'] = 0
df['cosHour'] = 0
df['sinWeekDay'] = 0
df['cosWeekDay'] = 0
def sinDayAndHour(s):
    s['hour'] = s['Momentum'].hour
    return s
    
df = df.apply(sinDayAndHour, 1)
df['sinHour'] = np.sin(2*np.pi*df['hour']/23)
df['sinWeekDay'] = np.sin(2*np.pi*df['weekDay']/6)
df['cosHour'] = np.cos(2*np.pi*df['hour']/23)
df['cosWeekDay'] = np.cos(2*np.pi*df['weekDay']/6)
df.head()
```

In order to train the neural network easily/faster, the weather columns will be normalized.

```python
df['T_noNorm'] = df['T']
def normalizeColumns(df, columns):
    for column in toNormalize:
        max_X = df[column].max()
        min_X = df[column].min()
        
        midrange = (max_X+min_X)/2
        Xrange = (max_X-min_X)
        
        df[column] = (df[column] - midrange) / (Xrange/2)
    return df

toNormalize = ['elektriciteit [kwh]', 'T', 'FH', 'U', 'N', 'Q']
df = normalizeColumns(df, toNormalize)
df.head()
```

```python
features = ['gas [m3]','elektriciteit [kwh]', 'T', 'FH', 'U', 'N', 'Q', 'sinHour', 'cosHour', 'sinWeekDay', 'cosWeekDay']
df[features].to_csv('filteredDatasets/'+building+'_NN_10F.csv', index=False)
```

## Remainders

## Outliers:

Now I have a good model, so I need to analyze the data and, if the prediction $p(x)$ is far from the actual data $a(x)$, I'll consider $x$ as outlier. 
* How to set this threshold?
* Will it work with the other buildings I have to analyze? (to verify this)

Let's try with a NN we trained before, with the method described [here](https://stats.stackexchange.com/questions/78079/confidence-interval-of-rmse).

```python
rr.tail()
```

```python
from scipy import stats
rr['outlier'] = 0
#ci = scipy.stats.chisquare(rr['predicted'], rr['actual'])
n = firstYear['gas [m3]'].count()
n = rr['actual'].count()
c1, c2 = stats.chi2.ppf([0.025,1-0.025],n)
l = sqrt(n/c2)*9.26
r = sqrt(n/c1)*9.26

import ml_metrics as metrics
err = metrics.rmse(rr['actual'].values, rr['predicted'].values)

print l, " ", r, " err:", err, " ", c1, " ", c2
```

```python
def contro(s):
    if abs(s['predicted']-s['actual']) > r*2.5:
        s['out2'] = 100
    else:
        s['out2'] = 0
    return s

rr = rr.apply(contro,1)
```
