# Dataset analysis

There are different factors that affect the energy needs of buildings. They can be divided into two groups: physical enviromental parameters (like outdoor Temperature <b>T</b>, solar radiation <b>Q</b>, wind speed <b>FH</b>) and artificial design parameters (like transparency ratio, orientation, thermo-physical properties).
I have a dataset composed by the hourly natural gas consumption of many building, the weather data of the place and the electric consumption of the same building.<br />
<b>TO DO:</b> find gas consumption outliers.

```python
import pandas as pd
import numpy as np
building = '740-NTH' #'882-WBW' #
df = pd.read_csv('datasets/'+building+'_elektriciteit_withWeather_clean.csv')
from datetime import date
df['Momentum'] = df['Momentum'].astype('datetime64')
df = df.set_index('Momentum', drop=False)

# There is an implicit conversion of timezone. I need to fix it.
df.index = df.index + pd.DateOffset(hours=1)
del df['Momentum']
df['Momentum'] = df.index

df.head()
```

## Gas behaviour
