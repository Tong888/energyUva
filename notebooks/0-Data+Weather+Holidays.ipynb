## Holidays implementation

Let's implement the weekDay in the dataset. The weekDay is a code from 0 to 6, where 0 represents Monday and 6 represents Sunday. All the dutch holidays are represented as Sunday, with code 6. (The holidays have a strange behaviour)

```python
import datetime

df['weekDay'] = 0
holidays = pd.read_csv('filteredDatasets/holidays.csv').values

def mergeWithHolidays(s):
    dateString = "%04d-%02d-%02d" % (s['Momentum'].year, s['Momentum'].month, s['Momentum'].day)
    weekday = 6
    if dateString not in holidays:
        weekday = s['Momentum'].weekday()
    s['weekDay'] = weekday
    return s

df = df.apply(mergeWithHolidays, 1)
df.head()
```

## Weather forecasts implementation

Now it's time to include in the dataset the weather informations. The columns will be explained later

```python
ff = pd.read_csv('datasets/KNMI_weatherdata[2008-2014].csv', sep=';')
columns = ff.columns
columns = [x.strip() for x in columns]
ff.columns = columns
ff = ff[['YYYYMMDD', 'HH', 'FH', 'T', 'U', 'N', 'Q']]
ff[['YYYYMMDD']] = ff[['YYYYMMDD']].astype(str)
ff[['U']] = ff[['U']].astype(float)
#HH is from 1 to 24
ff['HH'] = ff['HH']-1
#humidity is in percentage
ff.head()
```

```python
df['T'] = 0
df['FH'] = 0
df['U'] = 0
df['N'] = 0
df['Q'] = 0
def mergeWithForecasts(s):
    dateString = "%04d%02d%02d" % (s['Momentum'].year, s['Momentum'].month, s['Momentum'].day)
    selected = ff[(ff.YYYYMMDD == dateString) & (ff.HH == s['Momentum'].hour)]
    
    s['T'] = selected['T'].values[0]
    s['FH'] = selected['FH'].values[0]
    #percentual
    s['U'] = selected['U'].values[0]/100
    s['N'] = selected['N'].values[0]
    s['Q'] = selected['Q'].values[0]
    
    return s

df = df.apply(mergeWithForecasts, 1)
df.head()
```
